```python
def char_to_int(char):
    value = ord(char)-96
    if char == '.' : value = 27
    if char == '?' : value = 28
    
    return value
        
def word_to_int(word):
    word_value = []
    for char in word:
        word_value.append(char_to_int(char))
    return word_value

def line_to_list(line):
    words = line.split()
    the_list = []
    for word in words:
        the_list.append(word_to_int(word))
    return the_list

def num_to_list(the_num):
    a = the_num
    b = []
    while a/29 > 0:
        b.append(a%29)
        a = int(a/29)
    return b[::-1]

def list_to_word(line):
    word = ""
    for number in line:
        if number == 27:
            word += '.'
        elif number == 28:
            word += '?'
        else :
            word += chr(number+96)
    return word
```
